### Functions and if statements

```python
def check_score(score):
  """
  This  is a multiline comment.
  This a function header.
  check_score is a function that takes the score checks whether the student has passed the exam, by checking if the score is above 40
  """
  if score >= 40:
    print("The student has passed")
  else:
    print("The student has not passed")
  return

check_score(39)
check_score(99)

for s in scores:
  print(f"s={s},")
  check_score(s)    # function call, it will go to the function and when it encounters a return, it will come back to this function
```

### NUMPY array operations

```python
add_arr = arr2 + arr3
add_arr                                                     # It does element-wise addition

sub_arr =  arr2 - arr3
sub_arr                                                     # It does element-wise subtraction and it is true for other operations as well

mul_arr = arr2 * arr3
mul_arr                                                     # It does element-wise multiplication

div_arr = arr2 / arr3
div_arr                                                     # It does element-wise division
```
